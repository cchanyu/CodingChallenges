**9. What is normalization and name all normalization forms?<br>**
Normalization in ML is to center the data to normalize it<br>
Normalization in data science is to organize the data, so it's more structured and efficient<br>

1NF: First Normal Form - each column in table contains only 1 value<br>
2NF: Second Normal Form - Eliminate Partial Dependency, ex. Golf, 5000 <- we already know golf costs 5000<br>
3NF: Third Normal Form - Eliminate transitive dependency. ex. Sport, Price, Instructor, college <- college can be in separate table with instructor<br>
BCNF: Boyce-Codd Normal Form - each determinant be a candidate key, eliminating all anomalies<br>
4NF: Fourth Normal Form - no multi-valued dependency<br>
5NF: Fifth Normal Form - decompose table into smaller specific table based on its dependency<br>
